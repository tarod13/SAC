```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
rewards_hopper_1 = np.loadtxt('hopper/h1/mean_rewards_1.txt')
rewards_hopper_2 = np.loadtxt('hopper/h2/mean_rewards_2.txt')
rewards_hopper_3 = np.loadtxt('hopper/h3/mean_rewards_hopper_3.txt')
rewards_cheetah_1 = np.loadtxt('cheetah/c1/mean_rewards_1.txt')
rewards_pendulum_1 = np.loadtxt('pendulum/p1/mean_rewards_1.txt')
rewards_pendulum_2 = np.loadtxt('pendulum/p2/mean_rewards_2.txt')
rewards_pendulum_3 = np.loadtxt('pendulum/p3/mean_rewards_3.txt')
rewards_pendulum_4 = np.loadtxt('pendulum/p4/mean_rewards_4.txt')
rewards_pendulum_5 = np.loadtxt('pendulum/p5/mean_rewards_5.txt')
rewards_pendulum_6 = np.loadtxt('pendulum/p6/mean_rewards_pendulum_6.txt')
rewards_pendulum_7 = np.loadtxt('pendulum/p7/mean_rewards_pendulum_7.txt')
rewards_pendulum_8 = np.loadtxt('pendulum/p8/mean_rewards_pendulum_8.txt')
rewards_pendulum_9 = np.loadtxt('pendulum/p9/mean_rewards_pendulum_9.txt')
rewards_pendulum_DDPG_1 = np.loadtxt('pendulum/DDPG/mean_rewards_2.txt')
rewards_pendulum_DDPG_2 = np.loadtxt('pendulum/DDPG/mean_rewards_3.txt')
rewards_pendulum_DDPG_3 = np.loadtxt('pendulum/DDPG/mean_rewards_4.txt')
rewards_pendulum_DDPG_4 = np.loadtxt('pendulum/DDPG/mean_rewards_5.txt')
rewards_pendulum_DDPG_5 = np.loadtxt('pendulum/DDPG/mean_rewards_6.txt')
rewards_swimmer_2 = np.loadtxt('swimmer/s2/mean_rewards_swimmer_2.txt')
```

```python
rewards_hopper_1 = np.array(rewards_hopper_1)
rewards_hopper_2 = np.array(rewards_hopper_2)
rewards_hopper_3 = np.array(rewards_hopper_3)
rewards_cheetah_1 = np.array(rewards_cheetah_1)
rewards_pendulum_1 = np.array(rewards_pendulum_1)
rewards_pendulum_2 = np.array(rewards_pendulum_2)
rewards_pendulum_3 = np.array(rewards_pendulum_3)
rewards_pendulum_4 = np.array(rewards_pendulum_4)
rewards_pendulum_5 = np.array(rewards_pendulum_5)
rewards_pendulum_6 = np.array(rewards_pendulum_6)
rewards_pendulum_7 = np.array(rewards_pendulum_7)
rewards_pendulum_8 = np.array(rewards_pendulum_8)
rewards_pendulum_9 = np.array(rewards_pendulum_9)
rewards_pendulum_DDPG_1 = np.array(rewards_pendulum_DDPG_1)
rewards_pendulum_DDPG_2 = np.array(rewards_pendulum_DDPG_2)
rewards_pendulum_DDPG_3 = np.array(rewards_pendulum_DDPG_3)
rewards_pendulum_DDPG_4 = np.array(rewards_pendulum_DDPG_4)
rewards_pendulum_DDPG_5 = np.array(rewards_pendulum_DDPG_5)
rewards_swimmer_2 = np.array(rewards_swimmer_2)
```

```python
rewards_hopper_1 = rewards_hopper_1.reshape(-1)
rewards_hopper_2 = rewards_hopper_2.reshape(-1)
rewards_hopper_3 = rewards_hopper_3.reshape(-1)
rewards_cheetah_1 = rewards_cheetah_1.reshape(-1)
rewards_pendulum_1 = rewards_pendulum_1.reshape(-1)
rewards_pendulum_2 = rewards_pendulum_2.reshape(-1)
rewards_pendulum_3 = rewards_pendulum_3.reshape(-1)
rewards_pendulum_4 = rewards_pendulum_4.reshape(-1)
rewards_pendulum_5 = rewards_pendulum_5.reshape(-1)
rewards_pendulum_6 = rewards_pendulum_6.reshape(-1)
rewards_pendulum_7 = rewards_pendulum_7.reshape(-1)
rewards_pendulum_8 = rewards_pendulum_8.reshape(-1)
rewards_pendulum_9 = rewards_pendulum_9.reshape(-1)
rewards_pendulum_DDPG_1 = rewards_pendulum_DDPG_1.reshape(-1)
rewards_pendulum_DDPG_2 = rewards_pendulum_DDPG_2.reshape(-1)
rewards_pendulum_DDPG_3 = rewards_pendulum_DDPG_3.reshape(-1)
rewards_pendulum_DDPG_4 = rewards_pendulum_DDPG_4.reshape(-1)
rewards_pendulum_DDPG_5 = rewards_pendulum_DDPG_5.reshape(-1)
rewards_swimmer_2 = rewards_swimmer_2.reshape(-1)
```

```python
steps = np.linspace(1,1250,1250)/1000
```

```python
plt.figure(figsize=(6.5,4))
#plt.subplot(2,2,1)
plt.plot(steps, rewards_hopper_1*1000, label='1', linewidth=0.7)
plt.plot(steps[:len(rewards_hopper_2)], rewards_hopper_2*1000, 'firebrick', label='5', linewidth=0.7)
plt.plot(steps[:len(rewards_hopper_3)], rewards_hopper_3*1000, 'orange', label='25', linewidth=0.7)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0), useMathText=True)
plt.grid()
plt.legend(fontsize='large')
plt.xlabel(r'Million steps')
plt.ylabel(r'Average return')
plt.tight_layout()
plt.savefig('reward_scales.jpg', dpi=300)
plt.show()
plt.close()
```

```python
plt.figure()
#plt.plot(steps[:len(rewards_pendulum_1)]/2*100, rewards_pendulum_1*500, linewidth=0.95)
plt.plot(steps[:len(rewards_pendulum_2)]/2*100, rewards_pendulum_2*500, linewidth=0.95)
plt.plot(steps[:len(rewards_pendulum_3)]/2*100, rewards_pendulum_3*500, linewidth=0.95)
plt.plot(steps[:len(rewards_pendulum_4)]/2*100, rewards_pendulum_4*500, linewidth=0.95)
plt.plot(steps[:len(rewards_pendulum_5)]/2*100, rewards_pendulum_5*500, linewidth=0.95)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0), useMathText=True)
plt.grid()
plt.xlabel(r'Ten thousand steps')
plt.ylabel(r'Average return')
plt.show()
plt.close()
```

```python
rewards_pendulum = (rewards_pendulum_2[0:80]+rewards_pendulum_6[0:80]+rewards_pendulum_7[0:80]+rewards_pendulum_8[0:80]+rewards_pendulum_9[0:80])/5
std_dev_pendulum = ((rewards_pendulum_2[0:80]-rewards_pendulum)**2 + (rewards_pendulum_6[0:80]-rewards_pendulum)**2+ (rewards_pendulum_7[0:80]-rewards_pendulum)**2+ (rewards_pendulum_8[0:80]-rewards_pendulum)**2+ (rewards_pendulum_9[0:80]-rewards_pendulum)**2)**0.5/(5**0.5)
```

```python
rewards_pendulum_DDPG = (rewards_pendulum_DDPG_1[0:80]+rewards_pendulum_DDPG_2[0:80] + rewards_pendulum_DDPG_3[0:80] + rewards_pendulum_DDPG_4[0:80] + rewards_pendulum_DDPG_5[0:80])/5
std_dev_pendulum_DDPG = ((rewards_pendulum_DDPG_1[0:80]-rewards_pendulum_DDPG)**2 + (rewards_pendulum_DDPG_2[0:80]-rewards_pendulum_DDPG)**2 + (rewards_pendulum_DDPG_3[0:80]-rewards_pendulum_DDPG)**2+ (rewards_pendulum_DDPG_4[0:80]-rewards_pendulum_DDPG)**2+ (rewards_pendulum_DDPG_5[0:80]-rewards_pendulum_DDPG)**2)**0.5/(5**0.5)
```

```python
fig = plt.figure(figsize=(7,7))
fig.subplots_adjust(left=0.03, right=0.97, hspace=0.1, wspace=0.32)
plt.subplot(2,2,1)
plt.plot(steps[:len(rewards_hopper_3)], rewards_hopper_3*1000, linewidth=0.85)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0), useMathText=True)
plt.grid()
plt.title(r'Hopper-v2')
plt.xlabel(r'Million steps')
plt.ylabel(r'Average return')
plt.subplot(2,2,2)
plt.plot(steps[:len(rewards_cheetah_1)], rewards_cheetah_1*1000, linewidth=0.85)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0), useMathText=True)
plt.grid()
plt.title(r'Half-Cheetah-v2')
plt.xlabel(r'Million steps')
plt.ylabel(r'Average return')
plt.subplot(2,2,3)
plt.plot(steps[:len(rewards_pendulum_DDPG)]/2*100, rewards_pendulum_DDPG*500, linewidth=1.0, label='DDPG')
plt.fill_between(steps[:len(rewards_pendulum_DDPG)]/2*100, (rewards_pendulum_DDPG-std_dev_pendulum_DDPG)*500, (rewards_pendulum_DDPG+std_dev_pendulum_DDPG)*500, alpha=0.5)
plt.plot(steps[:len(rewards_pendulum)]/2*100, rewards_pendulum*500, linewidth=1.0, label='SAC')
plt.fill_between(steps[:len(rewards_pendulum)]/2*100, (rewards_pendulum-std_dev_pendulum)*500, (rewards_pendulum+std_dev_pendulum)*500, alpha=0.5)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0), useMathText=True)
plt.grid()
plt.legend(fontsize='medium')
plt.title(r'Pendulum-v0')
plt.xlabel(r'Ten thousand steps')
plt.ylabel(r'Average return')
plt.subplot(2,2,4)
plt.plot(steps[:len(rewards_swimmer_2)], rewards_swimmer_2*1000, linewidth=0.85)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0), useMathText=True)
plt.grid()
plt.title(r'Swimmer-v2')
plt.xlabel(r'Million steps')
plt.ylabel(r'Average return')
plt.tight_layout()
plt.savefig('average_returns.jpg', dpi=300)
plt.show()
plt.close()
```
